```python
import matplotlib.pyplot as plt
import sklearn
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = datasets.load_digits()
data.keys()
```

```python
X = data['data']
y = data['target']
xtr, xte, ytr, yte = train_test_split(X, y, test_size = 0.33, random_state = 101)
```

```python
def draw_sample_label(X,y, ypred = None):
    X = X[:20]
    y = y[:20]
    plt.subplots(4,5, figsize = (20,16))
    for i in range (len(X)):
        plt.subplot(4,5,i+1)
        plt.imshow(X[i].reshape(8,8), cmap = 'grey')
        if ypred is None: 
            plt.title(f'y={y[i]}')
        else :
            plt.title(f'y={y[i]} yp = {ypred[i]}')
draw_sample_label(X,y)
```

```python
clf = LogisticRegression(max_iter=10000)
clf.fit(xtr, ytr)
```

```python
ypred = clf.predict(xte)
```

```python
ypred
```

```python
(yte != ypred).sum() / len (yte)
```

```python
mask = yte != ypred 
draw_sample_label(xte[mask], yte[mask], ypred[mask])
```
